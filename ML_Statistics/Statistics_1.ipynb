```python
a=data['pclass'].value_counts()
```

```python
a
```

```python
per=(a/a.sum()*100).values
per
```

```python
c=[]
cf=0
for i in per:
    cf=cf+i
    c.append(cf)
```

```python
c
```

```python
import matplotlib.pyplot as plt

plt.plot(c,marker='D')
plt.show()
```
