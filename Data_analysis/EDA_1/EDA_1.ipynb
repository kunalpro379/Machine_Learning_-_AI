```python
df_contry=pd.read_excel('./ZomatoDataSet/Country-Code.xlsx')
df_contry.head()
```

```python
df.merge(df,df_contry, on='Country Code', how='left')
```

```python
merged_df = df.merge(df_contry, on='Country Code', how='left')
merged_df.head()
```

```python
#to check data types
print(merged_df.dtypes)
```

```python
merged_df.Country.value_counts()
```

```python
country_names=merged_df.Country.value_counts().index
```

```python
country_val=merged_df.Country.value_counts().values
```

```python
plt.pie(country_val, labels=country_names)
```

```python
#top 3 contries
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

```python
ratings=merged_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'})
```

```python
ratings
```

# conclusions/Obs.
1. when rating  is between 4.5 to 4.9----> excellent
2. when rating are between 4 to 3.4 --> very good
3. when rating are between 3.5 to 3.4 ----> good
4. when rating is between 3 to 3.4 ----> average
5. when rating is between 2.5 to 2.9 ----> average
6. when rating is between 2 to 2.4 ----> average
