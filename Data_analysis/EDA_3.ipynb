```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('train.csv')
```

```python
print(dataset.shape)
```

```python
dataset.head()
```

```python
#missing values
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]   #list compression
for feature in features_with_na: 
    print(feature, np.round(dataset[feature].isnull().mean(),4),'missing values%')
```

```python
#missing values
for feature in features_with_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(), 1,0)# if null value transform it as 1
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
print("id of houses {}".format(len(dataset.Id)))
```

```python
numerical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes!='O' ]# list compresison
print('number of numerical variables ', numerical_feature)
dataset[numerical_feature].head()
```

```python
year_feature=[feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature ]
year_feature
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('house price vs YearSold')
```

```python
year_feature
```

```python
for feature in year_feature:
    if feature != 'YrSold':
        data=dataset.copy()
        data[feature]=(data['YrSold']-data[feature])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
    
```

```python
descrete_feature=[feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in year_feature+['id']]
print("Descretee Variables Count : {}".format(len(descrete_feature)))
```

```python
descrete_feature
```

```python
dataset[descrete_feature].head()
```

```python
for feature in descrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
continuous_feature=[feature for feature in numerical_feature if feature not in descrete_feature+year_feature+['id']]
format(len(continuous_feature))
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])#log normal transformation
        data['SalePrice']=np.log(data["SalePrice"])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

```python
#Outliers
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
```
