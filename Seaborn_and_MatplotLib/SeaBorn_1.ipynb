```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.relplot(x='total_bill', y='tip',kind='scatter', data=data)
```

```python
sns.relplot(x='total_bill', y='tip',kind='line', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',style='sex', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',style='sex', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',style='sex',size='size',data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',style='sex',size='size',sizes=(20,300),data=data)
```
