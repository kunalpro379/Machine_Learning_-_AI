```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('fivethirtyeight')
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.catplot(x='day', y='tip', kind='strip', data=tips)
```

```python
sns.catplot(x='day', y='tip', kind='strip', jitter=20,data=tips)  #jitter
```

```python
sns.catplot(x='day', y='tip', kind='strip', jitter=0,data=tips)  #jitter
```
